### Exercise 2. 

**Initialize all the weights and all biases to a common value (for instance 0). Try to train
the network either with GD or SGD. What happens? Why?**

As it can be seen from the plot below, the accuracy of the Gradient Descent model with all weights and biases intialized to zero performs very poorly. Infact, the accuracy is same in each Epoch as if only certain data is being correctly classified. This is due to the fact that, all the weights are zero which means all the gradients in all directions are same resulting in neurons learning same features in every iteration. Additionally, as per the standard literature, if we initialize the network with a constant weight then there will be a problem of exploading gradient (high initial value) and vanishing gradient (low value initialization).
